```python
df1=pd.DataFrame({'ID':[1,2,3,4],'class':[9,10,11,12]})
df1
```

# Pandas merge connects  columns or indexes in 
Dataframe based on one or more keys

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
df2
```

```python
pd.merge(df1,df2,on='ID')#only same ID Will throw
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
pd.merge(df1,df2,on='ID',how='outer',indicator=True)
```
