# fillna()

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Industry':'none','Growth':0})
```

```python
df.fillna(method='ffill')##fill the just above value in the place of na value
```

```python
df.fillna(method='pad')##fill the just above value in the place of na value
```

```python
df.fillna(method='bfill')#fill the just below value in the place of na value
```

```python
df.fillna(method='ffill',axis=0)##fill the just above value in the place of na value
```

```python
df.fillna(method='ffill',axis=1)##fill the just left side columns corresponding value in the place of na value
```

```python
df.fillna(method='bfill',axis=1)##fill the just right side columns corresponding value in the place of na value
```

```python
df.fillna(0,limit=1)#replace just one value which appears first null value with 0 rest null remain the same
```

```python
df.fillna(0,limit=2)#replace just two value which appears first null value with 0 rest null remain the same
```

```python
df.fillna(method='ffill',limit=1)###fill the just above value in the place of na value but only one one null value will be filed
```

```python
df.fillna(5,inplace=True)#fill 5 inplace of null
df
```
