```python
##Want to extract only first five rows from the data
```

```python
df=pd.read_csv('Fortune_10.csv',nrows=5)
df
```

```python
##want to extract only first two columns from the data
```

```python
df=pd.read_csv('Fortune_10.csv',usecols=[0,1])
df
```

```python
#want to extract specific column using column index from the data
```

```python
df=pd.read_csv('Fortune_10.csv',usecols=[2,4,7])
df
```

```python
###want to extract specific column using column index  and skip some rows from the data 
```

```python
df=pd.read_csv('Fortune_10.csv',usecols=[2,4,7],skiprows=[1,2,3])
df
```

```python
#give the columns name as your choice while extracting data
```

```python
df=pd.read_csv('Fortune_10.csv',names=['A','B','C','D','E','F','G','H'])
df
```
